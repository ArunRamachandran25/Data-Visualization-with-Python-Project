```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
Topics_DataFrame = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/Topic_Survey_Assignment.csv", index_col = 0)
Topics_DataFrame
```

```python
Topics_DataFrame.sort_values(by = "Very interested", ascending = False, inplace = True)
Topics_DataFrame
```

```python
cols = ["Very interested", "Somewhat interested", "Not interested"]
Topics_DataFrame[cols] = Topics_DataFrame[cols].div(Topics_DataFrame[cols].sum(axis = 1), axis = 0).multiply(100)
# Topics_DataFrame[cols] = (Topics_DataFrame[cols] / 2233) * 100

Topics_DataFrame
```

```python
ax = Topics_DataFrame.plot(kind = 'bar',
                           figsize = (20, 8),
                           color=['#5cb85c', '#5bc0de', '#d9534f'],
                           width = 0.8)

ax.set_title("Percentage of Respondents' interest in Data Science Areas", fontsize = 16)
ax.set_xlabel("Different Data Science Topics", fontsize = 14)
ax.set_ylabel("Count for each Topic", fontsize = 14)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.legend(fontsize = 14)

plt.xticks(fontsize = 14)

for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height}', (x + width/2, y + height * 1.02), ha = 'center')
```

```python
Crime_Rate_Data = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/Police_Department_Incidents_-_Previous_Year__2016_.zip",index_col = False)
Crime_Rate_Data
```

```python
Crime_Columns = Crime_Rate_Data[['PdDistrict', 'Location']]

Counts = Crime_Rate_Data['PdDistrict'].value_counts()

Counts = pd.DataFrame(Counts)

Counts = Counts.reindex(['CENTRAL', 'NORTHERN', 'PARK', 'SOUTHERN', 'MISSION', 'TENDERLOIN', 'RICHMOND', 'TARAVAL', 'INGLESIDE', 'BAYVIEW'])

Counts = Counts.reset_index()

Counts.rename({'index' : 'Neighborhood', 'PdDistrict' : 'Count'}, axis = 'columns', inplace = True)

Counts
```
